```python
import statsmodels.api as sm

data =sm.datasets.co2.load_pandas().data
data.co2.interpolate(inplace=True)
data = data.resample('M').mean()

res = sm.tsa.x13_arima_select_order(data.co2, x12path='./x13as/x13as.exe')
print(res.order, res.sorder)
```

```python
results = sm.tsa.x13_arima_analysis(data.co2, x12path='./x13as/x13as.exe')

fig = results.plot()
fig.set_size_inches(12, 5)
fig.tight_layout()
```
